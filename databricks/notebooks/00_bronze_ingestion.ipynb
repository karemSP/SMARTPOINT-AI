```python
from pyspark.sql.types import StructType,StructField, StringType, IntegerType, DoubleType, BinaryType, TimestampType
from datetime import datetime
```

```python
spark.sql("create schema if not exists cvs")
```

###Bronz_table

```python
schema_bronz = StructType(
    [
        StructField("cv_id",StringType(),True),
        StructField("source",StringType(),True),
        StructField("raw_content",BinaryType(),True),
        StructField("file_name",StringType(),True),
        StructField("file_type",StringType(),True),
        StructField("ingested_at",TimestampType(),True),
        StructField("source_url",StringType(),True),
        StructField("minio_path",StringType(),True),
        StructField("checksum",StringType(),True)
        
    ]
)
```

```python
row_data = [
    ("uuid-001", "upload_minio", b"contenu1", "karem_ben_idriss.pdf", "pdf", datetime(2026, 4, 1, 9, 30, 0), "s3a://cvs/raw/karem_ben_idriss.pdf", "raw/karem_ben_idriss.pdf", "sha256-001"),
    ("uuid-002", "linkedin", b"contenu2", "ibrahima_ndiaye.docx", "docx", datetime(2026, 4, 2, 11, 15, 0), "s3a://cvs/raw/ibrahima_ndiaye.docx", "raw/ibrahima_ndiaye.docx", "sha256-002"),
    ("uuid-003", "indeed", b"contenu3", "kawthar_chaka.json", "json", datetime(2026, 4, 3, 14, 0, 0), "s3a://cvs/raw/kawthar_chaka.json", "raw/kawthar_chaka.json", "sha256-003")
]
```

```python
df_bronz = spark.createDataFrame(row_data, schema_bronz)
```
